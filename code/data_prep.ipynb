```python
#some columns are redundant, keep only the ones I want do simple version first
columns = [
    "Cycle start time",
    "Cycle end time",
    "Recovery score %",
    "Resting heart rate (bpm)",
    "Heart rate variability (ms)",
    "Day Strain",
    "Average HR (bpm)",
    "Respiratory rate (rpm)",
    "Asleep duration (min)",
    "Deep (SWS) duration (min)",
    "REM duration (min)"
]

data = data[columns]
data.shape #13 biofeedback features, 167 days
```

```python
data.tail()
```

```python
data.columns
```

```python
#shift columns: [Last night’s recovery + today’s strain = today’s data]. 
col_shift = [ #everything but day strain and cycle end time
    "Cycle start time",
    "Recovery score %",
    "Resting heart rate (bpm)",
    "Heart rate variability (ms)",
    "Average HR (bpm)",
    "Respiratory rate (rpm)",
    "Asleep duration (min)",
    "Deep (SWS) duration (min)",
    "REM duration (min)"
]

data[col_shift] = data[col_shift].shift(-1)
data.tail()
data.head()
```

```python
#drop last empty rows
data = data.dropna()
```

```python
data.isna().sum()
```

```python
data.tail()
data.columns
```

```python
#drop cycle start and replace cycle end time with date only
del data["Cycle start time"]
data["Date"] = pd.to_datetime(data["Cycle end time"]).dt.date
del data["Cycle end time"]
data.head()
```

```python
data.shape #number of examples, # of features (-1 for date, 8 features and 1 target variable)
#write out pre-standardized data table:
data.to_csv("feature_table.csv")
```
